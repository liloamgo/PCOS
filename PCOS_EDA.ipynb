```python
#It searches for the module initially in the local scope import. 
#The value returned by the function are then reflected in the output of the initial code.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#pandas function to read csv files and do operations on it
data = pd.read_csv("./PCOS_New.csv")
```

```python
#Plot pairwise relationships in a dataset.
PCOS = sns.pairplot(data, corner=True)
```

```python
#used to find the pairwise correlation of all columns in the dataframe.
corr_mat = data.corr()
print(corr_mat)
```

COMMENTS:
- numbers closer to -1 defining a strong negative correlation, 
- numbers closer to 1 defining a strong positive correlation and 
- numbers closer to 0 meaning little to no correlation.

```python
#Plot rectangular data as a color-encoded matrix.
fig, ax = plt.subplots(figsize=(16,14)) 
corr_data = sns.heatmap(data.corr(), annot=False, vmin=-1, vmax=1, cmap = "coolwarm", ax=ax)
print("Relationship between Attributes")
```

COMMENTS:
- numbers closer to -1 defining a strong negative correlation, 
- numbers closer to 1 defining a strong positive correlation and 
- numbers closer to 0 meaning little to no correlation.

```python
#plot that show relationship between numerical variable and one or more categorical variables
pcos_con = sns.catplot('PCOS (Y/N)', data=data, kind='count', aspect=1.5)
pcos_con.set_xlabels('PCOS Condition')

print("Patients with or without PCOS as a Condition")
```

NOTE: Patients with PCOS condition are represented by 1 while patients with no PCOS are represented by 0

```python
print('Where 0 = No PCOS; 1 = With PCOS')
#Counting the frequency of each label
pcos_count = data.groupby('PCOS (Y/N)')['PCOS (Y/N)'].count()

#Calculating the sum of the frequency
pcos_sum = pcos_count.sum()

#Normalizing the frequency
pcos_prct = pcos_count.apply(lambda x: 100*x / pcos_sum)

print(pcos_count)
print(pcos_prct)
```

FINDINGS: 363 (67%) Patients with no PCOS. 176 (33%) Patients with PCOS.

```python
#plot that show relationship between numerical variable and one or more categorical variables
age_yrs = sns.catplot('Age (yrs)', data=data, kind='count', aspect=2.5)
age_yrs.set_xlabels('Age')
print("Patients Ages")
```

```python
#Counting the frequency of each label
age_count = data.groupby('Age (yrs)')['Age (yrs)'].count()

#Calculating the sum of the frequency
age_sum = age_count.sum()

#Normalizing the frequency
age_prct = age_count.apply(lambda x: 100*x / age_sum)

print(age_count)
print(age_prct)
```
